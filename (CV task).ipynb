```python
import cv2 
from matplotlib import pyplot as plt 
import numpy as np 
```

```python
# read image , display it , convert it to gray  
img = cv2.imread(r'D:\ai test\240878704_112069901195206_8615529885798632174_n.jpg',0)
cv2.imshow('image',img)
cv2.waitKey()
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
#calculate histogram 
plt.hist(img.ravel())
```

```python
# min , max , variance
print("minumun number=",np.min(img))
print("max number=",np.max(img))
print("variance=",np.var(img))
```

```python
#image negative
def inverse (image):
    x =image.shape[0]
    y = image.shape[1]
    size = (x,y)
    image_n =np.zeros(size,np.uint8)
    image_n.fill(255)
    image_n = image_n - image
    
            
    return image_n    


img_n = inverse(img)
cv2.imshow('inverse',img_n)
cv2.waitKey()

    
```

```python
# contrast streching and contratiction 
norm_img1 = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
norm_img2 = cv2.normalize(img, None, alpha=0, beta=1.2, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)


norm_img1 = (255*norm_img1).astype(np.uint8)
norm_img2 = np.clip(norm_img2, 0, 1)
norm_img2 = (255*norm_img2).astype(np.uint8)


cv2.imwrite("streching.jpg",norm_img1)
cv2.imwrite("contratiction.jpg",norm_img2)


cv2.imshow('original',img)
cv2.imshow('streching',norm_img1)
cv2.imshow('contratiction',norm_img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#gray level slicing 
x,y = img.shape
   
grays = np.zeros((x,y),dtype = 'uint8')

min_range = 100

max_range = 115

for i in range(x):

    for j in range(y):

        if img[i,j]>min_range and img[i,j]<max_range: grays[i,j] = 255

        else: grays[i,j] = 0


cv2.imshow('sliced image', grays)
cv2.waitKey()
```
